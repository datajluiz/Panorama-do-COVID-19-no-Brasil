# Panorama do COVID-19 no Brasil

12 de março, data que infelizmente iniciou essa trágica trajetória de mortes por covid no país com a primeira vítima, uma mulher de 57 anos que morava em São Paulo ,  após  5 meses,  no dia 08 de agosto de 2020, registramos 100 mil mortes no Brasil.

seguindo em caminho contrário muitas cidades vivem a reabertura dos locais públicos e diminuindo o isolamento do povo dentro de casa, gerando  praias cheias, reabertura de bares / shopping e jogos de futebol. 

Impactados pela crise econômica , muitos voltaram ao trabalho presencial, com o pensamento de que o isolamento jamais teria sido uma boa opção, porém  para outros insanidade essa atitude.

Seguindo em segundo lugar com maior número de mortes, perdendo apenas para os EUA, o Brasil persiste na incerteza sobre os reais números de infectados e mortos pela covid-19, devido a várias vertentes e dados variados.

Miguel Nicolelis, neurocientista e coordenador voluntário do Comitê Científico de Combate ao Coronavírus do Consórcio Nordeste,  Paulo Lotufo, epidemiologista da USP (Universidade de São Paulo) e Domingos Alves, professor da Faculdade de Medicina da USP, falam sobre a falta de dados oficiais que permitam compreender como a pandemia se comportou até o momento e a eficácia das medidas de combate.

   **"Chegamos a 100 mil mortos e ainda não caiu a ficha do tamanho dessa tragédia. Quando vimos as quedas das torres gêmeas de Nova York, em 2001, ficamos muito chocados. O mesmo quando vemos a queda de aviões. Mas, curiosamente, a cada três dias caem duas torres, a cada dia caem três Boeings. Este número de mortos seria equivalente ao de uma guerra, e parece que ainda não caiu a ficha."** 
 
                                       *Miguel Nicolelis*

De acordo com o IBGE (Instituto Brasileiro de Geografia e Estatística), só 324 das 5.570 cidades no Brasil têm população maior do que 100 mil habitantes, cerca de 6% do total de municípios

"É como se você olhasse da sua janela de um bairro populoso de uma grande cidade e não visse ninguém em todos aqueles prédios. Se todos os mortos tivessem vivido próximo a você, Você estaria sozinho em um raio de 6,5 km", explica Nicolelis.

Para se ter um comparativo, com esses números o Brasil já superou o total de mortos da Gripe Espanhola e a Guerra do Paraguai.


Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## _*2.1 - Análisando os Dados*_

Esta etapa tem por objetivo criar uma consciência situacional inicial e permitir um entendimento de como os dados estão estruturados.

**Dicionário das variáveis**

      iso_code - códigos de três letras do país;
      ontinent - Continente;
      location - Localização geográfica;
      date - Data de observação;
      total_cases - Total de casos confirmados de COVID-19;
      new_cases - Novos casos confirmados de COVID-19;
      total_deaths - Total de mortes atribuídas a COVID-19;
      new_deaths - Novas mortes atribuídas a COVID-19;
      total_cases_per_million - Total de casos confirmados de COVID-19 por 1.000.000 de pessoas;
      new_cases_per_million - Novos casos confirmados de COVID-19 por 1.000.000 de pessoas;   
      total_deaths_per_million - Total de mortes atribuídas a COVID-19 por 1.000.000 de pessoas;   
      new_deaths_per_million - Novas mortes atribuídas a COVID-19 por 1.000.000 de pessoas;   
      total_tests - Total de testes para COVID-19;   
      new_tests - Novos testes para COVID-19;   
      new_tests_smoothed - Novos testes para COVID-19 (suavizado por 7 dias). Para países que não relatam dados de teste diariamente, presumimos que os testes mudaram igualmente diariamente em qualquer período em que nenhum dado foi relatado. Isso produz uma série completa de números diários, que é então calculada em média ao longo de uma janela contínua de 7 dias;   
      total_tests_per_thousand - Total de testes para COVID-19 por 1.000 pessoas;   
      new_tests_per_thousand - Novos testes para COVID-19 por 1.000 pessoas;
      new_tests_smoothed_per_thousand - Novos testes para COVID-19 (suavizado 7 dias) por 1.000 pessoas;   
      tests_per_case - Testes conduzidos por novo caso confirmado de COVID-19, dado como uma média contínua de 7 dias (este é o inverso de Positive_rate);   
      positive_rate - A parcela de testes COVID-19 que são positivos, dada como uma média contínua de 7 dias (este é o inverso de tests_per_case);   
      tests_units - Unidades usadas pelo local para relatar seus dados de test;   
      stringency_index - Índice de Restrição de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamentos de locais de trabalho e proibições de viagens, redimensionado para um valor de 0 a 100 (100 = resposta mais estrita);   
      population - População em 2020;   
      population_density - Número de pessoas dividido por área terrestre, medido em quilômetros quadrados, ano mais recente disponível;   
      median_age - Idade média da população, projeção da ONU para 2020;   
      aged_65_older - Parcela da população com 65 anos ou mais, ano mais recente disponível;   
      aged_70_older - Parcela da população com 70 anos ou mais em 2015;   
      gdp_per_capita - Produto interno bruto em paridade de poder de compra (dólares internacionais constantes de 2011), ano mais recente disponível;   
      extreme_poverty - Parcela da população que vive em pobreza extrema, ano mais recente disponível desde 2010;   
      cardiovasc_death_rate - Taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de mortes por 100.000 pessoas);   
      diabetes_prevalence - Prevalência de diabetes (% da população de 20 a 79 anos) em 2017;   
      female_smokers - Parcela de mulheres que fumam, ano mais recente disponível;   
      male_smokers - Proporção de homens que fumam, ano mais recente disponível;   
      handwashing_facilities - Parcela da população com instalações básicas de lavagem das mãos no local, ano mais recente disponível;   
      hospital_beds_per_thousand - Leitos hospitalares por 1.000 pessoas, ano mais recente disponível desde 2010;   
      life_expectancy - Expectativa de vida ao nascer em 2019.

Agora, vamos verificar os casos de morte.

```
df.loc[df.date == '2020-08-18'].sort_values(by="total_deaths", ascending=False)
```

Em relação às mortes, os países que lideravam aos 14 de agosto de 2020 eram:

Para melhorar a visualização, vamos plotar um gráfico de barras.

```
top_5_mortes = df.loc[df.date == '2020-08-18', ['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(15,10))

top_5_mortes.plot(x="location", y="total_deaths", kind="bar", ax=ax)

ax.set_title('Países com maior quantitativo de mortes por Covid - 19')
ax.set_xlabel('Países')
ax.set_ylabel('Quantidade de mortes por Covid - 19')

plt.tight_layout()
```

Como o vírus está progredindo?

Uma das maiores preocupações das organizações de saúde é a progressão do número de casos e mortes no mundo.

Uma das melhores formas de visualizarmos a evolução ao longo do tempo, é o gráfico de linhas. Por isso, vamos plotar um gráfico de linha para demonstrar e evolução de mortes no mundo.



```
# gráfico de linha para mortes no mundo

fig, ax = plt.subplots(figsize=(15,10))

df.loc[df.location == 'World', 'total_deaths'].reset_index(drop=True).plot(figsize=(10,8))

ax.set_title('Evolução de mortes de covid-19 no mundo')
ax.set_xlabel('Dias')
ax.set_ylabel('Mortes por covid')

plt.tight_layout()
```
